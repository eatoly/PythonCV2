```python
import numpy as np

import time
import cv2
```

```python
def brightPassFilter(image, thresh):
    
    imgx = image.shape[0]
    imgy = image.shape[1]
    
    #this is how you create a new np array!! Don't use '='
    edit = np.copy(image)
    
    for i in range(0, imgx):
        for j in range(0, imgy):
            if image[i][j][0] < thresh and image[i][j][1] < thresh and image[i][j][2] < thresh:
                edit[i][j] = 0
                
    return edit
    
def bloom(image, thresh, blurx, blury):
    
    imgBrightPass = brightPassFilter(image, thresh)
    
    imgBlur = cv2.blur(imgBrightPass, (blurx, blury))

    bloom = cv2.addWeighted(imgBlur, 1, img, 1, 0)
    
    return bloom


#start of program
img = cv2.imread('city.jpg')

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

for i in range(0,1):
    img = bloom(img, 150, 10, 10)

                
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()


    
```
