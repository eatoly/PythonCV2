```python
def imgSubtractHorz(image):
    
    edit = np.copy(image)

    for row in range(0, edit.shape[0]):
        for col in range(0, edit.shape[1] - 1):
            edit[row][col] = image[row][col] - image[row][col+1]
    
    return edit
    
    
def imgSubtractVert(image):
    
    edit = np.copy(image)

    for row in range(0, edit.shape[0]-1):
        for col in range(0, edit.shape[1]):
            edit[row][col] = image[row][col] - image[row+1][col]
    
    return edit
    
    
#start of program
img = cv2.imread('pics\scene.jpg')

edit = imgSubtractHorz(img)
edit2 = imgSubtractVert(img)

cv2.imwrite('sceneEdit1.png',edit)
cv2.imwrite('sceneEdit2.png',edit2)

cv2.imshow('edit', edit)
cv2.imshow('edit2', edit2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
